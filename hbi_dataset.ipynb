```python
#normalizing the data values
df4norm = (df4 - df4.mean()) / df4.std()
df4norm
```

```python
df4norm.isnull().sum().sort_values(ascending=False).head(20)
```

```python
df4norm = df4norm.drop(['Discomfort Location: ALL OVER', 'Discomfort Location: HEAD', 'Discomfort Location: CHEST', 
                        'Discomfort Location: BACK', 'Discomfort Location: HANDS', 'Discomfort Location: LEGS', 
                        'Discomfort Location: FEET', 'Discomfort Location: ARMS '], axis=1)
df4norm
```

```python
df4norm.isnull().sum().sort_values(ascending=False)
```

```python
#applying PCA
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

principalComponents = pca.fit_transform(df4norm)

principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])
principalDf.index = df4norm.index
principalDf
```

```python
#applying clustering on PCA
from matplotlib import pyplot
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.cluster

km = sklearn.cluster.KMeans()
labels = km.fit_predict(principalDf)
labelsDf = pd.DataFrame(labels, columns=['cluster'])
labelsDf.index = principalDf.index
labelsDf
```

```python
dataid = pd.concat([principalDf, labelsDf], axis=1)
dataid
```

```python
plt.scatter(dataid['principal component 1'], dataid['principal component 2'], c=dataid['cluster'], cmap='rainbow')
```

```python
for clusters in range(8):
    print(dataid[dataid['cluster'] == clusters].index.drop_duplicates())
```

```python
dataid.to_csv('dataid.csv')
```

```python
#use the elbow method to find optimal number of clusters 

from sklearn.cluster import KMeans

inertia = []
K = range(1, 10)
  
for k in K:
    km = KMeans(n_clusters=k)
    km.fit(dataid)
    inertia.append(km.inertia_)
```

```python
#plot the elbow
plt.plot(K, inertia, 'bx-')
plt.xlabel('K')
plt.ylabel('Inertia')
plt.show()
```

```python
km2 = sklearn.cluster.KMeans(n_clusters=5)
labels2 = km2.fit_predict(principalDf)
labelsDf2 = pd.DataFrame(labels2, columns=['cluster'])
labelsDf2.index = principalDf.index
labelsDf2
```

```python
dataid2 = pd.concat([principalDf, labelsDf2], axis=1)
dataid2
```

```python
plt.scatter(dataid2['principal component 1'], dataid2['principal component 2'], c=dataid2['cluster'], cmap='rainbow')
```

```python
dataid2.to_csv('dataid2.csv')
```
